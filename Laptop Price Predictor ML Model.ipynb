```python
laptops_price['Memory'] = laptops_price['Memory'].astype(str).replace('\.0', '', regex=True)

laptops_price["Memory"] = laptops_price["Memory"].str.replace('GB', '')
laptops_price["Memory"] = laptops_price["Memory"].str.replace('TB', '000')
new = laptops_price["Memory"].str.split("+", n = 1, expand = True)

laptops_price["first"]= new[0]
laptops_price["first"]=laptops_price["first"].str.strip()

laptops_price["second"]= new[1]

laptops_price["Layer1HDD"] = laptops_price["first"].apply(lambda x: 1 if "HDD" in x else 0)
laptops_price["Layer1SSD"] = laptops_price["first"].apply(lambda x: 1 if "SSD" in x else 0)
laptops_price["Layer1Hybrid"] = laptops_price["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
laptops_price["Layer1Flash_Storage"] = laptops_price["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

laptops_price['first'] = laptops_price['first'].str.replace(r'\D', '')

laptops_price["second"].fillna("0", inplace = True)

laptops_price["Layer2HDD"] = laptops_price["second"].apply(lambda x: 1 if "HDD" in x else 0)
laptops_price["Layer2SSD"] = laptops_price["second"].apply(lambda x: 1 if "SSD" in x else 0)
laptops_price["Layer2Hybrid"] = laptops_price["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
laptops_price["Layer2Flash_Storage"] = laptops_price["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

laptops_price['second'] = laptops_price['second'].str.replace(r'\D', '')

laptops_price["first"] = laptops_price["first"].astype(int)
laptops_price["second"] = laptops_price["second"].astype(int)

laptops_price["HDD"]=(laptops_price["first"]*laptops_price["Layer1HDD"]+laptops_price["second"]*laptops_price["Layer2HDD"])
laptops_price["SSD"]=(laptops_price["first"]*laptops_price["Layer1SSD"]+laptops_price["second"]*laptops_price["Layer2SSD"])
laptops_price["Hybrid"]=(laptops_price["first"]*laptops_price["Layer1Hybrid"]+laptops_price["second"]*laptops_price["Layer2Hybrid"])
laptops_price["Flash_Storage"]=(laptops_price["first"]*laptops_price["Layer1Flash_Storage"]+laptops_price["second"]*laptops_price["Layer2Flash_Storage"])

laptops_price.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

```python
X = laptops_price.drop(columns=['Price'])
y = np.log(laptops_price['Price'])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15, random_state=2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.svm import SVR
```

```python

from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

# Linear Regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tanf' ,OneHotEncoder(sparse=False, drop='first'),[0,1,7,10,11])
],remainder = 'passthrough')

step2 = LinearRegression()

pipe =Pipeline([
    ('step1', step1),
    ('step2', step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('Score ', r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

# Random Forest

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
